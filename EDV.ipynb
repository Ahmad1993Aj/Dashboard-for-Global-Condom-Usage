```python
for col in df.columns:
    if df[col].dtypes != "object":
        df[col].plot(
            kind = "box",
            subplots = True,
            figsize = (15, 5)
        )
    plt.title(f"Ausreißer in den numerischen Spalten --> {col}")
    plt.show()
```

```python
for col in df.columns:
    if df[col].dtypes == "object":
        print("++++****++++****++++\n")
        print(f"--> {col} <--")
        print(df[col].value_counts())
        print("\n++++****++++****++++\n")
```

```python
num_col = [col for col in df.columns if df[col].dtypes !="object"]
num_col
```

```python
df[num_col].mean()
```

```python
### Histogramm with matplotlib: ### 
for col in num_col:
    plt.hist(
        df[col],
        bins=20,
        density=True
    )
    plt.title(f"The Histogramm from the nummerical cols---> {col} <---")
    plt.show()
```

```python
sns.histplot(df[num_col]);
```

```python
### Histogramm with Seaborn: ### 
for col in num_col:
    sns.histplot(df[col], bins=20,
                 kde=True)
    plt.title(f"The Histogramm from the nummerical cols---> {col} <---")
    plt.show()
```

```python
cat_col = [col for col in df.columns if df[col].dtypes =="object"]
cat_col
```

```python
for col in cat_col:
    counts = df[col].value_counts().sort_values(ascending=False)
    total = len(df[col])
    plt.figure(figsize=(10, 5))
    sns.barplot(
        x = df[col].value_counts().index,
        y = df[col].value_counts().values,
        palette="viridis",
        legend=True,
    )
    plt.title(f"Verteilung von '{col}' (Gesamt: {total} Einträge)", pad=20, fontsize=14)
    plt.xlabel(col, labelpad=10, fontsize=12)
    plt.ylabel("Anzahl", labelpad=10, fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.show()
```
